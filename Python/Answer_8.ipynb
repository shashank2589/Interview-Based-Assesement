```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('westworld.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

## ● Get all the overall ratings for each season and using plots compare the ratings for all the seasons, like season 1 ratings, season 2, and so on.

```python
df['season'].unique()
```

```python
# Group the data by season and calculate the average overall ratings

average_ratings = df.groupby('season')['rating.average'].mean().reset_index()
```

```python
average_ratings
```

```python
# Line Plot of the ratings for all the seasons

plt.figure(figsize=(10, 6))
plt.plot(average_ratings['season'], average_ratings['rating.average'], marker='o')
plt.xlabel('Season')
plt.ylabel('Overall Rating')
plt.title('Overall Ratings by Season')
plt.xticks(average_ratings['season'])
plt.grid(True)
plt.show()
```

## ● Get all the episode names, whose average rating is more than 8 for every season

```python
# Filter the episodes by average rating of more than 8 for every season
rating_more_than_8 = df[df['rating.average'] > 8]
```

```python
rating_more_than_8
```

```python
# Create a bar plot of average ratings by season

plt.figure(figsize=(15, 7))
sns.barplot(x=rating_more_than_8['name'], y=rating_more_than_8['rating.average'], hue=rating_more_than_8['season'])
plt.xlabel('Episode Names')
plt.ylabel('Average Rating')
plt.title('Average Ratings (>8) by Season')
plt.show()
```

## ● Get all the episode names that aired before May 2019

```python
# Convert airdate column from string to datetime format 
df['airdate'] = pd.to_datetime(df['airdate'])
```

```python
# Filter out the episodes aired before May 2019
episodes_before_may_2019 = df[df['airdate'] < pd.to_datetime('2019-05-01')]
```

```python
episodes_before_may_2019.tail()
```

```python
# Create a histogram of episode counts by airdate
plt.figure(figsize=(10, 6))
sns.histplot(data=episodes_before_may_2019, x='airdate', bins='auto', kde=True, hue='season')
plt.xlabel('Airdate')
plt.ylabel('Episode Count')
plt.title('Episode Counts by Airdate (Before May 2019)')
plt.xticks(rotation=45)
plt.show()
```

## ● Get the episode name from each season with the highest and lowest rating

```python
# Group the episodes by season and find the episode name with the highest rating
highest_rating_episodes = df.loc[df.groupby('season')['rating.average'].idxmax()]

# Group the episodes by season and find the episode name with the lowest rating
lowest_rating_episodes = df.loc[df.groupby('season')['rating.average'].idxmin()]
```

```python
# find the highest rating per season
highest_rating = df.groupby('season')['rating.average'].max()
highest_rating
```

```python
# find the lowest rating per season
lowest_rating = df.groupby('season')['rating.average'].min()
lowest_rating
```

```python
highest_rating_episodes
```

```python
lowest_rating_episodes
```

```python
# Create scatter plots
plt.figure(figsize=(15, 7))

# Scatter plot for highest rated episodes
plt.scatter(highest_rating_episodes['name'], highest_rating_episodes['rating.average'],
            color='green', label='Highest Rated Episodes', marker='s', s=80)

# Scatter plot for lowest rated episodes
plt.scatter(lowest_rating_episodes['name'], lowest_rating_episodes['rating.average'],
            color='red', label='Lowest Rated Episodes', marker='^', s=80)

# Set plot title and labels
plt.title('Ratings of Highest and Lowest Rated Episodes')
plt.xlabel('Episode Name')
plt.ylabel('Average Rating')

# Add legend
plt.legend()

# Show the plot
plt.show()
```
