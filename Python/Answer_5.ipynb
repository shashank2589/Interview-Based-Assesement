```python
import requests
import pandas as pd

# Download the data from the provided link
result = requests.get("http://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes")
data = result.json()
```

```python
print(data.keys())
```

```python
# Create a DataFrame to store the structured data
# Use json_normalize to handle nested JSON and retrieve all the records and meta data
df = pd.json_normalize(data, record_path=['_embedded', 'episodes'], meta=['name', 'language'], meta_prefix = 'meta')
```

```python
# Display the first few rows of the DataFrame
df.head()
```

```python
# Convert the DataFrame to CSV format and save it to a file
df.to_csv("westworld.csv", index=False)
```
