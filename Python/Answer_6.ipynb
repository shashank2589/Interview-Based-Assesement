‚óè Get all Pokemons whose spawn rate is less than 5%

```python
# Filter the DataFrame based on spawn rate
spawn_rate = df[df['spawn_chance'] < 0.05]
```

```python
spawn_rate.head()
```

```python
# Plot the Pokemon with spawn rate less than 5%
plt.figure(figsize=(25, 8))
plt.bar(spawn_rate['name'], spawn_rate['spawn_chance'], color = "maroon")
plt.xlabel("Pokemon")
plt.ylabel("Spawn Rate")
plt.title("Pokemon with Spawn Rate Less Than 5%", fontsize = 20)
plt.xticks(rotation=90)
plt.show()
```

‚óè Get all Pokemons that have less than 4 weaknesses

```python
# Filter the DataFrame to get Pokemon with less than 4 weaknesses

# The lambda function is applied to the 'weaknesses' column, splitting the string by comma then checked for its length
weaknesses_df = df[df['weaknesses'].apply(lambda x: len(x.split(','))) < 4]

# Count the number of Pokemon in each category
weaknesses_count = weaknesses_df['weaknesses'].apply(lambda x: len(x.split(","))).value_counts()

```

```python
weaknesses_df.head()
```

```python
weaknesses_count
```

```python
# Plot the Pokemon that have less than 4 weaknesses

plt.figure(figsize=(8, 8))
plt.pie(weaknesses_count, labels=weaknesses_count.index, autopct='%1.2f%%')
plt.title('Pokemon with Less Than 4 Weaknesses', fontsize = 18)
plt.show()
```

‚óè Get all Pokemons that have no multipliers at all

```python
# Filter the DataFrame to get Pokemon with no multipliers
non_multipliers_df = df[df['multipliers'].isnull() == True]

# Filter the DataFrame to get Pokemon with multipliers
multipliers_df = df[df['multipliers'].isnull() == False]

# Count the length of multipliers and non multipliers
non_multiplier_count = len(non_multipliers_df)
multiplier_count = len(multipliers_df)
```

```python
non_multipliers_df.head()
```

```python
# Create labels for the pie chart
labels = ['Non-Multipliers', 'Multipliers']

# Create sizes for the pie chart
sizes = [non_multiplier_count, multiplier_count]

# Plot the Pokemon that have nomultipliers at all
plt.figure(figsize=(6, 8))
plt.pie(sizes, labels=labels,explode=(0.1, 0), autopct='%1.2f%%', startangle=90)
plt.title('Distribution of Pokemon Multipliers', fontsize = 18)
plt.show()
```

‚óè Get all Pokemons that do not have more than 2 evolutions

```python
# Filter the Pokemon with not more than 2 evolutions

# The lambda function is applied to the 'next_evolution' column, splitting the string by comma then checked for its length
evolution_df = df[df['next_evolution'].fillna('').apply(lambda x : len(x.split(','))) <= 2]
evolution_df.head()
```

```python
# Count the number of Pokemon in each category
pokemon_count = evolution_df['next_evolution'].fillna('').apply(lambda x: len(x.split(','))).value_counts().sort_index()

# Plot the pie chart of Pokemon that do not have more than 2 evolutions
plt.figure(figsize=(6, 8))
plt.pie(pokemon_count.values, labels=pokemon_count.index, autopct='%1.2f%%', explode=(0.1, 0), colors=['red', 'white'], wedgeprops={"edgecolor":"black"})
plt.title('Pokemon with Not More Than 2 Evolutions', fontsize = 15)
plt.show()
```

‚óè Get all Pokemons whose spawn time is less than 300 seconds.

Note - spawn time format is "05:32‚Äù, so assume ‚Äúminute: second‚Äù format and perform the analysis.

```python
# Filter the Pokemon whose spawn time is less than 300 seconds (5 minutes)

# The lambda function is applied to the 'spawn_time' column, splitting the string by colon then multiply minute with 60 to convert min to seconds and add it to seconds then check weather its less than 300 seconds
spawn_time_df = df[df['spawn_time'].fillna('00:00').apply(lambda x: int(x.split(':')[0]) * 60 + int(x.split(':')[1])) < 300]
```

```python
spawn_time_df.head()
```

```python
# Plot the countplot of Pokemon with Spawn Time Less Than 5 Minutes (300 seconds)

plt.figure(figsize=(20, 8))
sns.countplot(data=spawn_time_df, x='spawn_time')
plt.xlabel('Spawn Time (less than 5 minutes)')
plt.ylabel('Count')
plt.title('Pokemon with Spawn Time Less Than 5 Minutes', fontsize = 15)
plt.xticks(rotation=90)
plt.show()
```
